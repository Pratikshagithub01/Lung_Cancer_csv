```python
import numpy as np
import pandas as pd
from sklearn.cluster import AffinityPropagation
n = 4
data = np.random.randn(20)
index_nan = np.random.choice(data.size, n, replace=False)
index_inf = np.random.choice(data.size, n, replace=False)
data[index_nan]=np.nan
data[index_inf]=np.inf
print(data)
```

```python
#accuracy on training data
X_train_predicition = lr.predict(X_train)
training_data_accuracy = accuracy_score(y_train, X_train_predicition)
print(training_data_accuracy)
```

```python
#accuracy on test data
X_test_predicition = lr.predict(X_test)
test_data_accuracy = accuracy_score(y_test, X_test_predicition)
print(test_data_accuracy)
```
